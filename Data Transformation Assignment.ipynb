```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Load dataset
df = pd.read_csv('C:\\Users\\DivyaTejAtluri\\Downloads\\employee_productivity_dataset.csv')

# Display first rows
df.head()
```

Q1. Handle Missing Values (Basic Cleaning)

Fill missing values using the following:

Age → Median

Salary → Mean

Hours_Worked_Per_Week → Median

Performance_Score → Mean

Display the dataset after handling missing values

```python
df['Age'].fillna(df['Age'].median())
df['Salary'].fillna(df['Salary'].mean())
df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median())
df['Performance_Score'].fillna(df['Performance_Score'].mean())

df.head()
```

Q2. Label Encoding

Convert these columns into numeric values using Label Encoding:

Gender

Department

Show the updated columns

```python
le = LabelEncoder()

df['Gender'] = le.fit_transform(df['Gender'])
df['Department'] = le.fit_transform(df['Department'])

df[['Gender', 'Department']].head()
```

Q3. One-Hot Encoding

Apply One-Hot Encoding on:

Work_Mode

Location

Display the dataset and check how many new columns are created.

```python
df = pd.get_dummies(df, columns=['Work_Mode', 'Location'])

df.head()
print("Total columns after encoding:", df.shape[1])
```

Q4. Normalization (Min-Max Scaling)

Normalize the following columns:

Salary

Hours_Worked_Per_Week

Use MinMaxScaler and display results.

```python
scaler = MinMaxScaler()

df[['Salary', 'Hours_Worked_Per_Week']] = scaler.fit_transform(
    df[['Salary', 'Hours_Worked_Per_Week']]
)

df[['Salary', 'Hours_Worked_Per_Week']].head()
```

Q5. Standardization (Scaling)

Apply StandardScaler on:

Age

Projects_Completed

Display the transformed values.

```python
scaler_std = StandardScaler()

df[['Age', 'Projects_Completed']] = scaler_std.fit_transform(
    df[['Age', 'Projects_Completed']]
)

df[['Age', 'Projects_Completed']].head()
```

Q6. Compare Scaling Methods

Apply both:

MinMaxScaler

StandardScaler

On the Salary column.

Show both results side by side.

```python
salary = df[['Salary']]

minmax = MinMaxScaler().fit_transform(salary)
standard = StandardScaler().fit_transform(salary)

comparison = pd.DataFrame({
    'Original': salary.values.flatten(),
    'MinMax': minmax.flatten(),
    'Standardized': standard.flatten()
})

comparison.head()
```

Q7. Build Preprocessing Pipeline

Create a pipeline that:

Applies encoding to categorical columns

Applies scaling to numerical columns

Use:

ColumnTransformer

Pipeline

```python
num_cols = ['Age', 'Salary', 'Hours_Worked_Per_Week', 'Projects_Completed']
cat_cols = ['Gender', 'Department', 'Work_Mode', 'Location']

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(), cat_cols)
])

pipeline = Pipeline([
    ('preprocessing', preprocessor)
])

pipeline.named_steps['preprocessing']

pipeline
```

Q8. Apply Pipeline

Apply the pipeline on the dataset.

Display:

Transformed dataset

Shape of final dataset 



```python
df_original = pd.read_csv('C:\\Users\\DivyaTejAtluri\\Downloads\\employee_productivity_dataset.csv')

transformed_data = pipeline.fit_transform(df_original)

print("Shape of transformed data:", transformed_data.shape)
```
